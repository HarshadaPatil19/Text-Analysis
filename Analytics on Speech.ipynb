# Lemmatization

```python
from textblob import TextBlob, Word
import pattern
from pattern.en import lemma, lexeme
```

```python
def lemmatize_with_postage(text):
    sent = TextBlob(text)
    tag_dict = {"J":'j',
               "N":'n',
               "V": 'v',
               "R":'r'}
    words_and_tags = [(w, tag_dict.get(pos[0], 'n')) for w, pos in sent.tags]
    lemmatized_list = [wd.lemmatize(tag) for wd, tag in words_and_tags]
    return " ".join(lemmatized_list)
```

```python
Lemmatized_sentences =" ".join([lemma(wd) for wd in text.split()])
Lemmatized_sentences
```

# Regex or Regular Expression
A regular expression is a sequence of characters that define a search pattern.

```python
import re
```

```python
pattern = r"[^\w]"
Regular_Expression =  re.sub(pattern, " ", Lemmatized_sentences)
Regular_Expression
```

# Stopwords

```python
# import the stopwords
from nltk.corpus import stopwords
stop_words = set(stopwords.words("english"))
stop_words
```

# 
First will do sentence tokenization and word tokenize of the regular expression and then remove the stopwords

```python
tokenized_speech = sent_tokenize(Lemmatized_sentences)
tokenized_speech
```

```python
tokenized_words = word_tokenize(Regular_Expression)
tokenized_words
```

```python
filtered_sent = []
for w in tokenized_words:
    if w not in stop_words:
        filtered_sent.append(w)
```

```python
print(filtered_sent)
print("Length: ", len(filtered_sent))
```
